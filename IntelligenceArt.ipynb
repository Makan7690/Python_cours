```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
#np.random.seed(0) ==random_state=0
X, y = make_regression(n_samples=100, n_features=1, noise=5, random_state=0)

model = LinearRegression()

model.fit(X, y)

plt.scatter(X, y) # afficher le nuage de points.
plt.plot(X,model.predict(X), color = "red") # afficher la courbe de notre model.
```

```python
model.score(X,y)
```

```python
model.predict([[1]])
```

```python
mse = mean_squared_error(y, model.predict(X))
r2 = r2_score(y, model.predict(X))
print(f"Erreur quadratique moyenne (MSE) : {mse:.2f}")
print(f"Score R^2 : {r2:.2f}")
```
